```python
import pandas as pd

# Assuming you have your DataFrame 'df'
df = pd.DataFrame({
    'Country': ['Switzerland', 'Chile', 'Japan', 'United States', 'Slovenia'],
    'Voting Age Population (%)': [40.0, 45.7, 52.0, 53.6, 54.1],
    'Registered Voters (%)': [49.1, 42.0, 52.7, 84.3, 51.7]
})

# Get the LLM's response as a string
llm_response = """
import pandas as pd

def convert_to_dict(df):
    # Identify the column for y-axis
    y_axis_col = None
    for col in df.columns:
        if 'country' in col.lower() or 'location' in col.lower():
            y_axis_col = col
            break

    # Identify the columns for x-axises
    x_axises_cols = []
    for col in df.columns:
        if df[col].dtype.kind in ['i', 'f'] and ('vote' in col.lower() or 'population' in col.lower()):
            x_axises_cols.append(col)

    # Create the dictionary
    data = {
        "y-axis": df[y_axis_col].tolist(),
        "x-axises": [df[col].tolist() for col in x_axises_cols]
    }

    return data
"""

# Execute the LLM's response
exec(llm_response)
```

```python

# Call the function from the LLM's response
data_dict = convert_to_dict(df)

# Print the resulting dictionary
print(data_dict)
```

```python
from openai import OpenAI
import os
from dotenv import load_dotenv
from typing import List
import json 

try:
    load_dotenv(r"/home/shamal/AgentPlotly/.env")
except FileNotFoundError:
    print("No .Env File found.Please add .env file")
```

```python
api_key = os.environ.get("OPENAI_KEY")
client = OpenAI(api_key=api_key)
```

```python
from pydantic import BaseModel
from typing import List


class DataSource(BaseModel):
    y_axis :str
    x_axises: List[str]
```

```python
tool_extract_yandx_columns = {
    "type" : "function" ,
    "function" : {
        "name" : "extract_x_and_y_columns_from_dataframe" , 
        "description" : "Given x and y columns extract them from the dataframe" ,
        "parameters" : DataSource.model_json_schema()
    }
}
```

```python
response = client.chat.completions.create(
    model="gpt-4" ,
    messages = [
        {
            "role" :"system" ,
            "content" : "you are tasked with identifying the appropriate x and y axies that are need to create a plot that is requested by the user. You are given all the columns present in the data source. You may choose any number of columns for the x axsis but only choose one column for the y axis."
        }, 
        {
            'role' : "user" ,
            "content": "create a scatterplot so show the relationship between countries and the voting population.lets make the coutry the y axis. The columns provided are - `country`: A list of strings representing the countries or locations. `voting_pop`: A list of numbers representing the percentage of estimated voting age population. `reg_voters`: A list of numbers representing the percentage of estimated registered voters."
        }
    ] , 
    tools = [tool_extract_yandx_columns] ,
    tool_choice = {
        "type" : "function" ,
        "function" : {"name":"extract_x_and_y_columns_from_dataframe"}
    }
)
```

```python
response_message = response.choices[0].message
response_message
print(DataSource.model_validate_json(response_message.tool_calls[0].function.arguments))
```
